### Log Distribution

3. np.log
4. np.log1p. It adds 1 to all values performing doing the log.
5. np.log: use when x is expected to be significantly greater than zero and not in the immediate vicinity of 1. np.log1p: use when x can be very small (close to zero).
6. np.exp(value) - 1 or use np.expm1(value)
